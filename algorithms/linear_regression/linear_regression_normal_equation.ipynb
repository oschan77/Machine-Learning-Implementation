```python
import numpy as np
```

```python
def linear_regression_normal_equation(X, y):
    bias = np.ones((X.shape[0], 1))
    X_with_bias = np.append(bias, X, axis=1)
    w = np.dot(np.linalg.pinv(np.dot(X_with_bias.T, X_with_bias)), np.dot(X_with_bias.T, y))

    print(f'X.shape: {X.shape}')
    print(f'y.shape: {y.shape}')
    print(f'w.shape: {w.shape}')
    print(f'w_pred: {w}')

if __name__== "__main__":
    n_samples = 1000
    n_features = 4

    X = np.random.randn(n_samples, n_features)

    y = np.random.randn(n_samples, 1) * 0.01
    for n in range(n_features):
        w_true = np.random.randint(1, 10)
        y += np.expand_dims(w_true * X[:,n], axis=1)
        print(f'w_true: {w_true}')

    linear_regression_normal_equation(X, y)
```

```python
# X: (num_training_samples, num_features)
# X_with_bias: (num_training_samples, num_features +1)
# y: (num_training_samples, 1)
# W: (num_features + 1, 1)
```
